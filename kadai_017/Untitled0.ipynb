```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print("名前:", self.name)
        print("年齢:", self.age)

    def is_adult(self):
        if self.age >= 20:
            return "大人です"
        else:
            return "大人ではないです"


# 複数人分のインスタンスを作成
person_list = [
    Human("佐藤さとこ", 28),
    Human("田中たろう", 17),
    Human("鈴木はな", 35)
]

# クラス外では「お願いするだけ」
for person in person_list:
    person.printinfo()
    print(person.is_adult())
    print("-----")
```
