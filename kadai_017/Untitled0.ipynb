```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(self.name, "is adult.")
        else:
            print(self.name, "is not adult.")


# 複数人分のインスタンスを作成
person_list = [
    Human("佐藤さとこ", 20),
    Human("田中たろう", 17),
    Human("鈴木はな", 35)
]

# クラス外ではメソッドを呼ぶだけ
for person in person_list:
    person.check_adult()
```
