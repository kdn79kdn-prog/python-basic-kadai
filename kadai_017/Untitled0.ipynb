```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print("名前:", self.name)
        print("年齢:", self.age)

person = Human("佐藤さとこ", 28)

person.printinfo()

if 20 < person.age:
    print("大人です")

if 20 > person.age:
    print("大人ではないです")
```
